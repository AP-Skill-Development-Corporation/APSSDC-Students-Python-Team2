#### Types of Arguments
- positional arguments
- keyword arguments
- default arguments
- arbitrary arguments/variable length arguments

```python
# positional arguments
# no of inputs should be equal to the no of arguments 
def multiple(a,b):
    print(a,b)
    return a*b

multiple(20,56)

```

```python
multiple(56,20)
```

```python
multiple(56,20,10)
```

```python
# keyword arguments
def multiple(a,b):
    print(a,b)
    return a*b

multiple(20,56)  #positional arguments
```

```python
multiple(b=56,a=20)   # keyword arguments
```

```python
# default arguments
def multiple(a,b=5):
    print(a,b)
    return a*b

multiple(20)  #default arguments
```

```python
multiple(20,30)
```

```python
# varible length arguments/arbitrary arguments
def func4(*names):
    for i in names:
        print(i,end="\t")
#names="au","lbrce","anits","gitam","srkr","gvpce","gmrit"
func4("au","lbrce","anits","gitam","srkr","gvpce","gmrit")
```

#### recursive function

```python

def function4(c):
    print("python")
    function4(c)
function4(c)
```

```python
#factorail problem 
#5!=5*4*3*2*1=120


def factorial(n):
    p=1
    for i in range(1,n+1):
        p=p*i
    print(p)
#factorial(5)
factorial(int(input()))
```

```python
def factorail(n):
    if n==1 or n==0:
        return 1
    else:
        return n*factorial(n-1)   # 5*factorial(4)
                                  # 5*4*factorial(3)
                                  # 5*4*3*factorial(2)
                                  # 5*4*3*2*factorial(1)
                                  # 5*4*3*2*1=120
factorial(0)
    
```

```python
#factors of a given number
#6--->1,2,3,4,5,6-->1,2,3,6
def factor(n):
    for i in range(1,n+1):
        if n%i==0:
            print(i,end=" ")
factor(6)
```

```python
# check whether the given number is prime or not
#2,3,5,7,11....
#2--1,2
def factcount(n):
    c=0
    for i in range(1,n+1):
        if n%i==0:
            c+=1
    return c
```
